```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/dats.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.replace('-', np.nan, inplace=True)
```

```python
# Fill missing values in categorical columns with "Unknown"
df['PRIORITY_1'].fillna('Unknown', inplace=True)
df['PRIORITY_2'].fillna('Unknown', inplace=True)
df['PRIORITY_3'].fillna('Unknown', inplace=True)
df['PRIORITY_4'].fillna('Unknown', inplace=True)
df['PRIORITY_5'].fillna('Unknown', inplace=True)
```

```python
df.head()
```

```python
# Perform one-hot encoding on the priority columns
df = pd.get_dummies(df, columns=['PRIORITY_1', 'PRIORITY_2', 'PRIORITY_3', 'PRIORITY_4', 'PRIORITY_5'])
```

```python
df.head()
```

```python
df.columns
```

```python


# Create a mask to identify rows where all priority columns are zero or "Unknown"
mask = (df[['PRIORITY_1_Adventure', 'PRIORITY_1_Food and Drinks', 'PRIORITY_1_History and Culture',
             'PRIORITY_1_Local Experiences', 'PRIORITY_1_Museum', 'PRIORITY_1_Religious', 'PRIORITY_1_Scenic',
             'PRIORITY_1_Shopping', 'PRIORITY_1_Shows and Concerts',
             'PRIORITY_2_Adventure', 'PRIORITY_2_Food and Drinks', 'PRIORITY_2_History and Culture',
             'PRIORITY_2_Local Experiences', 'PRIORITY_2_Museum', 'PRIORITY_2_Scenic', 'PRIORITY_2_Shopping',
             'PRIORITY_2_Shows and Concerts', 'PRIORITY_2_Unknown',
             'PRIORITY_3_Adventure', 'PRIORITY_3_Food and Drinks', 'PRIORITY_3_History and Culture',
             'PRIORITY_3_Local Experiences', 'PRIORITY_3_Religious', 'PRIORITY_3_Scenic', 'PRIORITY_3_Shopping',
             'PRIORITY_3_Shows and Concerts', 'PRIORITY_3_Unknown',
             'PRIORITY_4_Adventure', 'PRIORITY_4_History and Culture', 'PRIORITY_4_Local Experiences',
             'PRIORITY_4_Religious', 'PRIORITY_4_Scenic', 'PRIORITY_4_Shopping',
             'PRIORITY_4_Shows and Concerts', 'PRIORITY_4_Unknown',
             'PRIORITY_5_Adventure', 'PRIORITY_5_Food and Drinks', 'PRIORITY_5_Local Experiences',
             'PRIORITY_5_Religious', 'PRIORITY_5_Scenic', 'PRIORITY_5_Shopping', 'PRIORITY_5_Unknown']] == 0).all(axis=1)
# Create a new column "PRIORITY_Unknown" and set it to 1 for rows where all priority columns are zero or "Unknown"
df['PRIORITY_Unknown'] = mask.astype(int)

# Now, you have "PRIORITY_Unknown" as a separate column indicating the lowest priority
```

```python
X = df.drop(['PID', 'POIs'], axis=1)  # Features (excluding PID and POIs columns)
y = df['POIs']  # Target variable
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
y_encoded
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X, y_encoded, test_size=0.2, random_state=42)
```

```python
len(df.columns)
```

```python
number_of_features = len(df.columns) - 2  # Subtract 3 for the excluded columns
# Calculate the number of unique POIs to determine the number of classes
number_of_classes = len(df['POIs'].unique())
```

```python
import tensorflow as tf

model = tf.keras.Sequential([
    # Input layer with the shape of your feature vector
    tf.keras.layers.Input(shape=(number_of_features,)),
    # Hidden layers (add more as needed)
    tf.keras.layers.Dense(128, activation='relu'),
    # Output layer with softmax activation (number_of_classes is the number of unique POIs)
    tf.keras.layers.Dense(number_of_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train_encoded, epochs=100, batch_size=50)
```

```python


# Convert the user's priorities to a one-hot encoded feature vector
user_priorities = ['Adventure', 'Scenic', 'Unknown', 'Unknown', 'Unknown']

# Create a DataFrame with the same columns as your original dataset
user_feature_vector = pd.DataFrame(columns=X_train.columns)

# Initialize all columns to 0
user_feature_vector.loc[0] = 0

# Set the columns corresponding to the user's priorities to 1
for i, priority in enumerate(user_priorities):
    if priority != 'Unknown':
        user_feature_vector[f'PRIORITY_{i+1}_{priority}'] = 1

# Convert the user's features to a NumPy array
user_features = user_feature_vector.values

# Predict the probabilities of all POIs for the user
predictions = model.predict(user_features)

# Get the indices of the top N recommended POIs
N = 5  # Adjust this value to get a specific number of recommendations
top_indices = np.argsort(predictions[0])[::-1][:N]

# Decode the recommended POIs using the label encoder
recommended_pois = label_encoder.inverse_transform(top_indices)

# Print the recommended POIs
print("Recommended POIs:")
for poi in recommended_pois:
    print(poi)
```
